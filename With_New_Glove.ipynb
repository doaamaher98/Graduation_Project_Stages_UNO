```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X= sc.fit_transform(dataset4.drop(["Letter"],axis = 1))
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.25,random_state=42, stratify=y)
```

```python
from sklearn.svm import SVC
model = SVC(kernel='rbf',degree=3)
model.fit(X_train,y_train)

# Predicting the Test set results SVM
y_pred = model.predict(X_test)

#calculate the details SVM
print('train_score svcmodel', model.score(X_train,y_train))
print('test_score svcmodel',model.score(X_test,y_test))
# Making the Confusion Matrix
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test,y_pred)
print(cm)
```
